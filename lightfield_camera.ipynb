### Setting up Functions

```python
originX = 8
originY = 8
def calc_shift(imname, alpha):
    name = imname.split("-")

    fix = name[0].split("_")
    x, y = int(fix[2]), int(fix[1])
    print(x)
    print(y)

    x_move = x - originX
    y_move = originY - y
    res1 = (x, y)
    res2 = (alpha * y_move, alpha * x_move, 0)
    return res1, res2
```

```python

def implement_shift(img, sh):
    assert len(sh) == len(img.shape)
    res_img = np.zeros(img.shape)

    select_org = []
    select_res = []

    for s in sh:
        if s == 0:
            sel_org = slice(None)
            res_sel = slice(None)
        elif s > 0:
            sel_org= slice(s, None)
            res_sel = slice(None, -s)
        else:
            sel_org = slice(None, s)
            res_sel = slice(-s, None)

        select_org.append(sel_org)
        select_res.append(res_sel)
    res_img[select_org] = img[select_res]
    return res_img

def dist(x_1, y_1, x_2, y_2):
    return math.sqrt((x_2 - x_1 )**2 + (y_2 - y_1 )**2) 

```

```python
def calc_avg_img(fl, s, r, fl_name):
    fl_cwd = os.getcwd()
    direc_img = os.path.join(fl_cwd, fl)
    c, img_avg = 1, None

    list_dir = os.listdir(direc_img)
    for fl in list_dir:
        org_res, cor_res = calc_shift(fl, s)
        if "out_08_08" in fl:
            path = os.path.join(direc_img, fl)
            img_avg = skio.imread(path ) /255.
            break

    assert img_avg.all() != None

    for fl in list_dir:
        org_res, cor_res = calc_shift(fl, s)

        if "out_08_08" in fl:
            continue

        if dist(originX, originY, org_res[0], org_res[1]) <= r:
            path = os.path.join(direc_img, fl)
            res_img = skio.imread(path ) /255.
            res_img = implement_shift(res_img, cor_res)
            img_avg += res_img
            c += 1

    img_avg = img_avg /float(c)
    skio.imsave("{}_avg_img.png".format(fl_name), img_avg)
```

### Depth Refocusing

```python
def depthRefocusing(directory):
    
    ### Calculate the average image with shifting
    for i in range(0, 7):
        calc_avg_img(directory, i, 25, "./{}/depth_refocusing{}".format("processed_" + directory, i))

depthRefocusing("jewels")
```
